## Match momentum

Wykres pokazuje przewag? jednej z dru?yn w kolejnych fragmentach meczu. Dodatnie s?upki oznaczaj? momentum gospodarzy, ujemne momentum go?ci.

```python
import html
import numpy as np


def _team_name_from_match(row, column):
    value = row[column]
    if isinstance(value, dict):
        return value.get(f"{column}_name")
    return value


def _safe_series(frame, column, default=np.nan):
    if column in frame.columns:
        return frame[column]
    return pd.Series(default, index=frame.index)


def _x_coordinate(value):
    if isinstance(value, (list, tuple)) and value:
        return value[0]
    return np.nan


def build_match_momentum(events, match_info, window=5):
    goal_weight = 15.0
    shot_base_weight = 1.0
    shot_xg_weight = 6.0
    own_goal_weight = 5.0
    final_third_pass_weight = 0.35
    final_third_carry_weight = 0.25
    complete_dribble_weight = 0.25
    recovery_weight = 0.15

    momentum = events.copy()
    momentum["match_minute"] = momentum["minute"] + _safe_series(momentum, "second", 0).fillna(0) / 60
    momentum["shot_xg"] = _safe_series(momentum, "shot.statsbomb_xg", 0).fillna(0)
    momentum["location_x"] = _safe_series(momentum, "location").apply(_x_coordinate)
    momentum["carry_end_x"] = _safe_series(momentum, "carry.end_location").apply(_x_coordinate)
    momentum["pass_end_x"] = _safe_series(momentum, "pass.end_location").apply(_x_coordinate)

    completed_pass = _safe_series(momentum, "pass.outcome.name").isna()
    shot_goal = momentum["type.name"].eq("Shot") & _safe_series(momentum, "shot.outcome.name").eq("Goal")
    shot_score = np.where(
        momentum["type.name"].eq("Shot"),
        shot_base_weight + shot_xg_weight * momentum["shot_xg"],
        0,
    )
    shot_score = np.where(shot_goal, goal_weight, shot_score)
    pass_into_final_third = (
        momentum["type.name"].eq("Pass")
        & completed_pass
        & momentum["location_x"].lt(80)
        & momentum["pass_end_x"].ge(80)
    )
    carry_into_final_third = (
        momentum["type.name"].eq("Carry")
        & momentum["location_x"].lt(80)
        & momentum["carry_end_x"].ge(80)
    )
    complete_dribble = (
        momentum["type.name"].eq("Dribble")
        & _safe_series(momentum, "dribble.outcome.name").eq("Complete")
    )
    recovery = momentum["type.name"].eq("Ball Recovery")
    own_goal_for = momentum["type.name"].eq("Own Goal For")

    momentum["momentum_score"] = (
        shot_score
        + own_goal_for.astype(float) * own_goal_weight
        + pass_into_final_third.astype(float) * final_third_pass_weight
        + carry_into_final_third.astype(float) * final_third_carry_weight
        + complete_dribble.astype(float) * complete_dribble_weight
        + recovery.astype(float) * recovery_weight
    )
    momentum["time_bin"] = (momentum["match_minute"] // window * window).astype(int)

    teams = [
        _team_name_from_match(match_info, "home_team"),
        _team_name_from_match(match_info, "away_team"),
    ]
    bins = np.arange(0, int(np.ceil(momentum["match_minute"].max() / window) * window) + window, window)

    grouped = (
        momentum.groupby(["time_bin", "team.name"], as_index=False)["momentum_score"]
        .sum()
        .pivot(index="time_bin", columns="team.name", values="momentum_score")
        .reindex(index=bins, columns=teams)
        .fillna(0)
    )
    grouped["momentum_diff"] = grouped[teams[0]] - grouped[teams[1]]
    return grouped, teams, momentum


def _goal_events(momentum_events):
    shot_goals = momentum_events[
        momentum_events["type.name"].eq("Shot")
        & _safe_series(momentum_events, "shot.outcome.name").eq("Goal")
    ][["match_minute", "team.name"]].copy()
    own_goals = momentum_events[
        momentum_events["type.name"].eq("Own Goal For")
    ][["match_minute", "team.name"]].copy()
    return (
        pd.concat([shot_goals, own_goals], ignore_index=True)
        .sort_values("match_minute")
        .reset_index(drop=True)
    )


def _polyline(points):
    return " ".join(f"{x:.1f},{y:.1f}" for x, y in points)


def _step_points(minutes, values, scale_x, scale_y):
    points = [(scale_x(0), scale_y(0))]
    prev_y = scale_y(0)
    for minute, value in zip(minutes, values):
        x = scale_x(minute)
        y = scale_y(value)
        points.append((x, prev_y))
        points.append((x, y))
        prev_y = y
    return points


def plot_match_momentum(events, matches, match_id, window=5):
    match_info = matches.loc[matches["match_id"].eq(match_id)].iloc[0]
    momentum_table, teams, momentum_events = build_match_momentum(events, match_info, window=window)
    home_team, away_team = teams

    width, height = 1040, 650
    left, right = 78, 34
    top, mid_top, bottom = 64, 410, 600
    chart_w = width - left - right
    bar_h = 260
    line_h = 130
    max_minute = max(float(momentum_table.index.max() + window), 1)
    max_abs_momentum = max(float(momentum_table["momentum_diff"].abs().max()), 1)

    def sx(minute):
        return left + float(minute) / max_minute * chart_w

    zero_y = top + bar_h / 2

    def sy_momentum(value):
        return zero_y - float(value) / max_abs_momentum * (bar_h / 2 - 16)

    momentum_timeline = (
        momentum_events[momentum_events["momentum_score"].gt(0)]
        .groupby(["match_minute", "team.name"], as_index=False)["momentum_score"]
        .sum()
        .sort_values("match_minute")
    )
    momentum_series = {}
    max_cumulative_momentum = 1
    for team in teams:
        team_momentum = momentum_timeline[momentum_timeline["team.name"].eq(team)].copy()
        if team_momentum.empty:
            team_momentum["cum_momentum"] = []
        else:
            team_momentum["cum_momentum"] = team_momentum["momentum_score"].cumsum()
            max_cumulative_momentum = max(max_cumulative_momentum, float(team_momentum["cum_momentum"].max()))
        momentum_series[team] = team_momentum

    def sy_cumulative_momentum(value):
        return bottom - float(value) / max_cumulative_momentum * line_h

    home_color = "#2f80ed"
    away_color = "#eb5757"
    grid_color = "#d9dee7"
    text_color = "#20242a"
    parts = [
        f'<svg xmlns="http://www.w3.org/2000/svg" width="{width}" height="{height}" viewBox="0 0 {width} {height}">',
        '<rect width="100%" height="100%" fill="#ffffff"/>',
        f'<text x="{left}" y="34" font-family="Arial, sans-serif" font-size="22" font-weight="700" fill="{text_color}">Match momentum: {html.escape(home_team)} vs {html.escape(away_team)}</text>',
        f'<circle cx="{left}" cy="50" r="5" fill="{home_color}"/><text x="{left + 10}" y="54" font-family="Arial, sans-serif" font-size="13" fill="{text_color}">{html.escape(home_team)}</text>',
        f'<circle cx="{left + 150}" cy="50" r="5" fill="{away_color}"/><text x="{left + 160}" y="54" font-family="Arial, sans-serif" font-size="13" fill="{text_color}">{html.escape(away_team)}</text>',
    ]

    for factor in [-1, -0.5, 0, 0.5, 1]:
        y = sy_momentum(max_abs_momentum * factor)
        parts.append(f'<line x1="{left}" x2="{width - right}" y1="{y:.1f}" y2="{y:.1f}" stroke="{grid_color}" stroke-width="1"/>')
    parts.append(f'<line x1="{left}" x2="{width - right}" y1="{zero_y:.1f}" y2="{zero_y:.1f}" stroke="#222222" stroke-width="1.4"/>')
    parts.append(f'<text x="20" y="{zero_y:.1f}" font-family="Arial, sans-serif" font-size="12" fill="{text_color}" transform="rotate(-90 20 {zero_y:.1f})">Momentum</text>')

    bar_width = chart_w / len(momentum_table) * 0.82
    for minute, row in momentum_table.iterrows():
        value = row["momentum_diff"]
        x = sx(minute + window / 2) - bar_width / 2
        y = min(zero_y, sy_momentum(value))
        h = abs(sy_momentum(value) - zero_y)
        color = home_color if value >= 0 else away_color
        parts.append(f'<rect x="{x:.1f}" y="{y:.1f}" width="{bar_width:.1f}" height="{h:.1f}" fill="{color}" opacity="0.9"/>')

    goals = _goal_events(momentum_events)
    for _, goal in goals.iterrows():
        x = sx(goal["match_minute"])
        color = home_color if goal["team.name"] == home_team else away_color
        parts.append(f'<line x1="{x:.1f}" x2="{x:.1f}" y1="{top}" y2="{bottom}" stroke="{color}" stroke-width="1.5" stroke-dasharray="5 5" opacity="0.75"/>')
        parts.append(f'<text x="{x - 4:.1f}" y="{top + 18}" font-family="Arial, sans-serif" font-size="11" font-weight="700" fill="{color}" transform="rotate(-90 {x - 4:.1f} {top + 18})">GOL</text>')

    for minute in range(0, int(max_minute) + 1, 15):
        x = sx(minute)
        parts.append(f'<line x1="{x:.1f}" x2="{x:.1f}" y1="{top}" y2="{bottom + 8}" stroke="{grid_color}" stroke-width="1" opacity="0.55"/>')
        parts.append(f'<text x="{x:.1f}" y="{bottom + 30}" font-family="Arial, sans-serif" font-size="11" fill="{text_color}" text-anchor="middle">{minute}</text>')

    parts.append(f'<text x="20" y="{bottom - line_h / 2:.1f}" font-family="Arial, sans-serif" font-size="12" fill="{text_color}" transform="rotate(-90 20 {bottom - line_h / 2:.1f})">Skumulowane momentum</text>')
    parts.append(f'<line x1="{left}" x2="{width - right}" y1="{bottom}" y2="{bottom}" stroke="#222222" stroke-width="1.2"/>')
    parts.append(f'<line x1="{left}" x2="{left}" y1="{bottom - line_h}" y2="{bottom}" stroke="#222222" stroke-width="1.2"/>')
    for value in np.linspace(0, max_cumulative_momentum, 4):
        y = sy_cumulative_momentum(value)
        parts.append(f'<line x1="{left}" x2="{width - right}" y1="{y:.1f}" y2="{y:.1f}" stroke="{grid_color}" stroke-width="1"/>')
        parts.append(f'<text x="{left - 8}" y="{y + 4:.1f}" font-family="Arial, sans-serif" font-size="11" fill="{text_color}" text-anchor="end">{value:.1f}</text>')

    for team, color in [(home_team, home_color), (away_team, away_color)]:
        team_momentum = momentum_series[team]
        if team_momentum.empty:
            points = [(sx(0), sy_cumulative_momentum(0)), (sx(max_minute), sy_cumulative_momentum(0))]
        else:
            points = _step_points(team_momentum["match_minute"], team_momentum["cum_momentum"], sx, sy_cumulative_momentum)
            points.append((sx(max_minute), points[-1][1]))
        parts.append(f'<polyline points="{_polyline(points)}" fill="none" stroke="{color}" stroke-width="3" stroke-linejoin="round" stroke-linecap="round"/>')

    parts.append(f'<text x="{left + chart_w / 2:.1f}" y="{height - 12}" font-family="Arial, sans-serif" font-size="13" fill="{text_color}" text-anchor="middle">Minuta meczu</text>')
    parts.append('</svg>')
    return ''.join(parts), momentum_table


match_momentum_svg, match_momentum = plot_match_momentum(df, df_matches, match_id, window=5)
try:
    from IPython.display import SVG, display
    display(SVG(match_momentum_svg))
except ImportError:
    print(match_momentum_svg)

match_momentum.head()
```

## Match momentum - wersja matplotlib

Alternatywny wariant tego samego wykresu przygotowany w `matplotlib`.

```python
import matplotlib.pyplot as plt



def plot_match_momentum_matplotlib(events, matches, match_id, window=5):
    match_info = matches.loc[matches["match_id"].eq(match_id)].iloc[0]
    momentum_table, teams, momentum_events = build_match_momentum(events, match_info, window=window)
    home_team, away_team = teams

    home_score = match_info["home_score"]
    away_score = match_info["away_score"]

    momentum_timeline = (
        momentum_events[momentum_events["momentum_score"].gt(0)]
        .groupby(["match_minute", "team.name"], as_index=False)["momentum_score"]
        .sum()
        .sort_values("match_minute")
    )

    home_color = "#2f80ed"
    away_color = "#eb5757"
    colors = np.where(momentum_table["momentum_diff"].ge(0), home_color, away_color)

    fig, axes = plt.subplots(
        2,
        1,
        figsize=(14, 8),
        sharex=True,
        gridspec_kw={"height_ratios": [2, 1]},
    )

    axes[0].bar(
        momentum_table.index + window / 2,
        momentum_table["momentum_diff"],
        width=window * 0.85,
        color=colors,
        edgecolor="white",
        linewidth=0.8,
    )
    axes[0].axhline(0, color="#222222", linewidth=1)
    axes[0].set_ylabel("Momentum")
    axes[0].set_title(f"Match momentum: {home_team} {home_score} - {away_score} {away_team}")
    axes[0].grid(axis="y", alpha=0.25)

    max_minute = max(momentum_table.index) + window

    for team, color in [(home_team, home_color), (away_team, away_color)]:
        team_momentum = momentum_timeline[momentum_timeline["team.name"].eq(team)].copy()
        if team_momentum.empty:
            axes[1].step([0, max_minute], [0, 0], where="post", label=team, color=color, linewidth=2)
            continue

        team_momentum["cum_momentum"] = team_momentum["momentum_score"].cumsum()
        x_values = [0, *team_momentum["match_minute"].tolist(), max_minute]
        y_values = [0, *team_momentum["cum_momentum"].tolist(), team_momentum["cum_momentum"].iloc[-1]]
        axes[1].step(
            x_values,
            y_values,
            where="post",
            label=team,
            color=color,
            linewidth=2,
        )

    goals = _goal_events(momentum_events)
    top_ylim = axes[0].get_ylim()
    for _, goal in goals.iterrows():
        minute = goal["match_minute"]
        team = goal["team.name"]
        goal_color = home_color if team == home_team else away_color
        for ax in axes:
            ax.axvline(minute, color=goal_color, linestyle="--", linewidth=1, alpha=0.65)
        axes[0].text(
            minute,
            top_ylim[1] * 0.92,
            "GOL",
            rotation=90,
            color=goal_color,
            ha="right",
            va="top",
            fontsize=9,
            fontweight="bold",
        )

    axes[1].set_ylabel("Skumulowane momentum")
    axes[1].set_xlabel("Minuta meczu")
    axes[1].legend(loc="upper left", frameon=False)
    axes[1].grid(axis="y", alpha=0.25)
    axes[1].set_xlim(0, max_minute)

    plt.tight_layout()
    return fig, momentum_table


fig_matplotlib, match_momentum_matplotlib = plot_match_momentum_matplotlib(
    df,
    df_matches,
    match_id,
    window=5,
)
match_momentum_matplotlib.head(10)
```

## Skumulowane xG

Dodatkowy wykres pokazujący narastające xG obu drużyn w czasie meczu.

```python
import matplotlib.pyplot as plt


def plot_cumulative_xg(events, matches, match_id):
    match_info = matches.loc[matches["match_id"].eq(match_id)].iloc[0]
    home_team = _team_name_from_match(match_info, "home_team")
    away_team = _team_name_from_match(match_info, "away_team")
    home_score = match_info["home_score"]
    away_score = match_info["away_score"]
    teams = [home_team, away_team]

    xg_events = events.copy()
    xg_events["match_minute"] = xg_events["minute"] + _safe_series(xg_events, "second", 0).fillna(0) / 60
    xg_events["shot_xg"] = _safe_series(xg_events, "shot.statsbomb_xg", 0).fillna(0)
    max_minute = int(np.ceil(xg_events["match_minute"].max() / 5) * 5)

    shots = xg_events[xg_events["type.name"].eq("Shot")].copy()
    xg_timeline = (
        shots.groupby(["match_minute", "team.name"], as_index=False)["shot_xg"]
        .sum()
        .sort_values("match_minute")
    )

    colors = {
        home_team: "#2f80ed",
        away_team: "#eb5757",
    }

    fig, ax = plt.subplots(figsize=(14, 5))
    for team in teams:
        team_xg = xg_timeline[xg_timeline["team.name"].eq(team)].copy()
        if team_xg.empty:
            ax.step([0, max_minute], [0, 0], where="post", label=team, color=colors[team], linewidth=2.5)
            continue

        team_xg["cum_xg"] = team_xg["shot_xg"].cumsum()
        x_values = [0, *team_xg["match_minute"].tolist(), max_minute]
        y_values = [0, *team_xg["cum_xg"].tolist(), team_xg["cum_xg"].iloc[-1]]
        ax.step(
            x_values,
            y_values,
            where="post",
            label=f"{team}: {team_xg['cum_xg'].iloc[-1]:.2f} xG",
            color=colors[team],
            linewidth=2.5,
        )

    goals = _goal_events(build_match_momentum(events, match_info, window=5)[2])
    top_ylim = ax.get_ylim()
    for _, goal in goals.iterrows():
        minute = goal["match_minute"]
        team = goal["team.name"]
        goal_color = colors.get(team, "#222222")
        ax.axvline(minute, color=goal_color, linestyle="--", linewidth=1, alpha=0.65)
        ax.text(
            minute,
            top_ylim[1] * 0.94,
            "GOL",
            rotation=90,
            color=goal_color,
            ha="right",
            va="top",
            fontsize=9,
            fontweight="bold",
        )

    ax.set_title(f"Skumulowane xG: {home_team} {home_score} - {away_score} {away_team}")
    ax.set_xlabel("Minuta meczu")
    ax.set_ylabel("Skumulowane xG")
    ax.set_xlim(0, max_minute)
    ax.grid(axis="y", alpha=0.25)
    ax.legend(loc="upper left", frameon=False)

    plt.tight_layout()
    return fig


fig_cumulative_xg = plot_cumulative_xg(df, df_matches, match_id)
```
