```python
# World Cup 2018

import pandas as pd
import json
import os


with open("C:/Users/misie/Documents/UAM/MGR_MK/open-data-master/open-data-master/data/competitions.json", "r", encoding="utf-8") as f:
    competitions = json.load(f)

df_comp = pd.DataFrame(competitions)

# filtr World Cup 2018
wc = df_comp[
    (df_comp["competition_name"] == "FIFA World Cup") &
    (df_comp["season_name"] == "2018")
]

competition_id = wc.iloc[0]["competition_id"]
season_id = wc.iloc[0]["season_id"]

print("Competition ID:", competition_id)
print("Season ID:", season_id)
```

```python
match_id = df_matches.iloc[0]["match_id"]

print("Selected match_id:", match_id)
```

```python
events_path = f"C:/Users/misie/Documents/UAM/MGR_MK/open-data-master/open-data-master/data/events/{match_id}.json"

with open(events_path, "r", encoding="utf-8") as f:
    events = json.load(f)

df_events = pd.json_normalize(events)

print("Events loaded:", df_events.shape)

```

```python
df = df_events.copy()

df["is_pass"] = df["type.name"] == "Pass"
df["is_shot"] = df["type.name"] == "Shot"
df["is_carry"] = df["type.name"] == "Carry"
df["is_duel"] = df["type.name"].isin(["Duel", "Interception"])
```

```python


df["match_id"] = match_id  

player_match = df.groupby(
    ["match_id", "player.id", "player.name", "team.name"]
).agg(
    passes=("is_pass", "sum"),
    shots=("is_shot", "sum"),
    carries=("is_carry", "sum"),
    duels=("is_duel", "sum"),
    events=("type.name", "count")
).reset_index()

print("Player-match rows:", len(player_match))
```

```python
player_match
```

```python

minutes = df.groupby(
    ["match_id", "player.id"]
).agg(
    minutes=("minute", "max")
).reset_index()

player_match = player_match.merge(
    minutes,
    on=["match_id", "player.id"],
    how="left"
)
```

```python
for col in ["passes", "shots", "carries", "duels", "events"]:
    player_match[col + "_per90"] = (
        player_match[col] / player_match["minutes"] * 90
    )
```

```python

final_columns = [
    "match_id",
    "player.id",
    "player.name",
    "team.name",
    "minutes",
    "passes_per90",
    "shots_per90",
    "carries_per90",
    "duels_per90",
    "events_per90"
]

player_match_final = player_match[final_columns]

print("\nATASET")
print(player_match_final.head())
```

```python
match_duration = df["minute"].max()

minutes = (
    df.groupby(["match_id", "player.id"])
      .agg(
          first_minute=("minute", "min"),
          last_minute=("minute", "max")
      )
      .reset_index()
)

minutes["minutes_played"] = (
    minutes["last_minute"] - minutes["first_minute"] + 1
)

minutes["minutes_played"] = minutes["minutes_played"].clip(
    upper=match_duration
)

minutes
```

```python
player_match = player_match.drop(
    columns=["minutes"],
    errors="ignore"
)

player_match = player_match.merge(
    minutes[
        ["match_id", "player.id", "minutes_played"]
    ],
    on=["match_id", "player.id"],
    how="left"
)
```

```python
df["is_pass"] = df["type.name"] == "Pass"
df["is_shot"] = df["type.name"] == "Shot"
df["is_carry"] = df["type.name"] == "Carry"
```

```python
passes = df[df["type.name"] == "Pass"]
```

```python
df["completed_pass"] = (
    (df["type.name"] == "Pass") &
    (df["pass.outcome.name"].isna())
)
```

```python
passing_stats = df.groupby(
    ["match_id", "player.id"]
).agg(
    passes_completed=("completed_pass", "sum")
).reset_index()

player_match = player_match.merge(
    passing_stats,
    on=["match_id", "player.id"],
    how="left"
)

player_match["pass_accuracy"] = (
    player_match["passes_completed"] /
    player_match["passes"]
)
```

```python
df["shot_xg"] = df["shot.statsbomb_xg"].fillna(0)
```

```python
xg_stats = df.groupby(
    ["match_id", "player.id"]
).agg(
    total_xg=("shot_xg", "sum")
).reset_index()

player_match = player_match.merge(
    xg_stats,
    on=["match_id", "player.id"],
    how="left"
)
```

```python
df["is_pressure"] = (
    df["type.name"] == "Pressure"
)

pressure_stats = df.groupby(
    ["match_id", "player.id"]
).agg(
    pressures=("is_pressure", "sum")
).reset_index()

player_match = player_match.merge(
    pressure_stats,
    on=["match_id", "player.id"],
    how="left"
)
```

```python
df["is_interception"] = (
    df["type.name"] == "Interception"
)

interception_stats = df.groupby(
    ["match_id", "player.id"]
).agg(
    interceptions=("is_interception", "sum")
).reset_index()

player_match = player_match.merge(
    interception_stats,
    on=["match_id", "player.id"],
    how="left"
)
```

```python
per90_cols = [
    "passes",
    "shots",
    "carries",
    "duels",
    "events",
    "total_xg",
    "pressures",
    "interceptions"
]

for col in per90_cols:
    player_match[f"{col}_per90"] = (
        player_match[col] /
        player_match["minutes_played"] *
        90
    )
```

```python
feature_columns = [
    "match_id",
    "player.id",
    "player.name",
    "team.name",

    "minutes_played",

    "passes_per90",
    "shots_per90",
    "carries_per90",
    "duels_per90",

    "pressures_per90",
    "interceptions_per90",

    "total_xg_per90",

    "pass_accuracy",

    "events_per90"
]

player_match_final = player_match[
    feature_columns
]

```
