```python
import os
import pandas as pd
import json

DATA_PATH = r"C:\Users\misie\Documents\UAM\MGR_MK\open-data-master\open-data-master\data"

with open(f"{DATA_PATH}/matches/43/3.json", "r", encoding="utf-8") as f:
    matches = json.load(f)

df_matches = pd.DataFrame(matches)

world_cup_ids = set(
    str(x) for x in df_matches["match_id"]
)

events_dir = f"{DATA_PATH}/events"

for file in os.listdir(events_dir):

    match_id = file.replace(".json", "")

    if match_id not in world_cup_ids:
        os.remove(
            os.path.join(events_dir, file)
        )

print("Events cleaned")
```

```python
import os

lineups_dir = f"{DATA_PATH}/lineups"

removed = 0
kept = 0

for file in os.listdir(lineups_dir):

    if not file.endswith(".json"):
        continue

    match_id = file.replace(".json", "")

    if match_id in world_cup_ids:
        kept += 1
    else:
        os.remove(os.path.join(lineups_dir, file))
        removed += 1

print(f"Zostawiono: {kept}")
print(f"Usunięto: {removed}")
```

```python
import os
import shutil

matches_dir = f"{DATA_PATH}/matches"

for competition_folder in os.listdir(matches_dir):

    competition_path = os.path.join(
        matches_dir,
        competition_folder
    )

    if not os.path.isdir(competition_path):
        continue

    if competition_folder != "43":
        shutil.rmtree(competition_path)
        continue

    for file in os.listdir(competition_path):

        if file != "3.json":
            os.remove(
                os.path.join(
                    competition_path,
                    file
                )
            )

print("Matches cleaned")
```

```python
matches_path = f"{DATA_PATH}/matches/43/3.json"

with open(matches_path, "r", encoding="utf-8") as f:
    matches = json.load(f)

df_matches = pd.DataFrame(matches)
world_cup_ids = set(df_matches["match_id"].astype(str))

folders = {
    "events": f"{DATA_PATH}/events",
    "lineups": f"{DATA_PATH}/lineups",
    "three-sixty": f"{DATA_PATH}/three-sixty",
}

print("Liczba meczów World Cup 2018:", len(world_cup_ids))
print("-" * 50)

for name, folder in folders.items():
    existing_files = {
        file.replace(".json", "")
        for file in os.listdir(folder)
        if file.endswith(".json")
    }

    missing = world_cup_ids - existing_files
    extra = existing_files - world_cup_ids

    print(f"\nFolder: {name}")
    print("Plików JSON:", len(existing_files))
    print("Brakujące pliki:", len(missing))
    print("Nadmiarowe pliki:", len(extra))

    if missing:
        print("Missing:", sorted(list(missing))[:10])

    if extra:
        print("Extra:", sorted(list(extra))[:10])
```
